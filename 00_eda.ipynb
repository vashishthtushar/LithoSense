```python

# Load the clean CSV version
df = pd.read_csv("./data/raw/dataset-uci.csv")
print("Dataset loaded with shape:", df.shape)
df.head()
```

```python

# Basic info
df.info()
df.describe().T
```

```python
df['Gallstone Status'].value_counts(normalize=True).plot(kind="bar")
plt.title("Target distribution (Gallstone Status)")
plt.show()
```

```python

# Missingness heatmap
plt.figure(figsize=(12,6))
sns.heatmap(df.isna(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python

# Numeric distributions
df.hist(bins=30, figsize=(15,10))
plt.tight_layout()
plt.show()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Skewness check
df.skew(numeric_only=True).sort_values(ascending=False)
```

```python
#check missing values
df.isna().sum().sort_values(ascending=False)
```

```python
#check class distribution
df['Gallstone Status'].value_counts(normalize=True)
```
