```python
import pandas as pd
import numpy as np
from pathlib import Path
import pandas.api.types as ptypes
from IPython.display import Markdown

# Load dataset (adjust path if needed)
df = pd.read_csv("../data/raw/dataset-uci.csv")

# Prefill Descriptions & Units (extend as needed)
prefill = {
    "Gallstone Status": ("Categorical (0/1)", "Target variable: gallstones present (1) or absent (0)", "-"),
    "Age": ("Numeric", "Patient age", "Years"),
    "Gender": ("Categorical", "Patient sex (Male/Female)", "-"),
    "BMI": ("Numeric", "Body Mass Index", "kg/m²"),
    "Cholesterol": ("Numeric", "Serum cholesterol level", "mg/dL"),
    "Glucose": ("Numeric", "Fasting blood glucose level", "mg/dL"),
    "Triglycerides": ("Numeric", "Serum triglyceride level", "mg/dL"),
    "Blood Pressure": ("Numeric", "Systolic/diastolic blood pressure", "mmHg"),
    "Bioimpedance": ("Numeric", "Electrical bioimpedance (body composition)", "ohms"),
    "Comorbidity": ("Categorical", "Presence of other conditions (e.g. diabetes, hypertension)", "-"),
    "Smoking Status": ("Categorical", "Current/past smoker", "-"),
    "Alcohol Intake": ("Categorical", "Alcohol consumption (yes/no)", "-"),
    "Physical Activity": ("Categorical", "Level of physical activity", "-"),
}

rows = []
for col in df.columns:
    if col in prefill:
        col_type, desc, units = prefill[col]
    else:
        col_type = "Numeric" if ptypes.is_numeric_dtype(df[col]) else "Categorical"
        desc, units = "", ""

    # Notes field
    notes = ""
    if ptypes.is_numeric_dtype(df[col]):
        skew = df[col].skew()
        skew_type = "right-skewed" if skew > 0.5 else "left-skewed" if skew < -0.5 else "approx. symmetric"
        notes = f"Range: {df[col].min()}–{df[col].max()}, Skewness={skew:.2f} ({skew_type})"
    else:
        counts = df[col].value_counts(normalize=True).mul(100).round(1).astype(str) + "%"
        balance = ", ".join([f"{cls}: {pct}" for cls, pct in counts.items()])
        notes = f"Class balance → {balance}"

    rows.append(f"| {col} | {col_type} | {desc} | {units} | {notes} |")

# Build Markdown
content = (
    "# Data Dictionary — Gallstone Dataset\n\n"
    "This file documents each variable in the dataset, its meaning, and notes from EDA.\n\n"
    "> ✅ Note: No missing values were found in this dataset.\n\n"
    "| Column Name | Type | Description | Units | Notes (EDA findings) |\n"
    "|-------------|------|-------------|-------|-----------------------|\n"
    + "\n".join(rows) + "\n"
)

# Save file
Path("data_dictionary.md").write_text(content, encoding="utf-8")
print("✅ Auto-filled data_dictionary.md generated successfully!")

# Render inside Jupyter
Markdown(content)
```
